```python
import numpy as np
import math
def convolve(image: np.ndarray, kernel: np.ndarray) -> np.ndarray:
    """ Convolve an image with a kernel assuming zero padding of the image to handle the borders
    :param image: the image (either greyscale of the shape=(rows,cols) 
    :type 
    :param :
    type numpy.ndarray 
    kernel: the numpy.ndarray 
    :returns the kernel convolved 
    :rtype 
    """ 
    # Your # at numpy.ndarray code (shape=(kheight,kwidth); image or both image (of the same shape as the dimensions input to colour handle the borders shape=(rows,cols,channels)) odd) image) here. a reasonable You'll speed.
    # dimension of image
    (irows,icols,channels)=image.shape
    #dimension of template
    (trows,tcols)=kernel.shape
    # set resultant image to black
    finalImage=np.zeros(shape=(irows,icols,channels), dtype=float)
    #half of template row
    trhalf=math.floor(trows/2)
    #half of template columns
    tchalf=math.floor(tcols/2)
    inversetemplate=np.zeros(shape=(irows,icols,channels), dtype=float)
    for i in range(channels):
        for x in range(0,icols):
                for y in range(0,irows):
                    inversetemplate[x][y][i]=image[icols-x][irows-y][i]
   
    image=np.pad(image, ((tchalf, trhalf), (tchalf, trhalf), (0, 0)))
    for i in range(channels):
        for x in range(0,icols):
            for y in range(0,irows):

                finalImage[y][x][i]=np.sum(image[y:y+trows,x:x+tcols,i]*inversetemplate)
    
    return finalImage
    
```

```python
import math 
import numpy as np  

# from MyConvolution import convolve   
def myHybridImages(lowImage: np.ndarray, lowSigma: float, highImage: np.ndarray, highSigma: float) -> np.ndarray:  
    #Create hybrid images by combining a low-pass and high-pass filtered pair.         
#    :param lowImage: the image to low-pass filter (either greyscale shape=(rows,cols) or colour shape=(rows,cols,channels))     
#    :type numpy.ndarray  
#    :param lowSigma: the standard deviation of the Gaussian used for low-pass filtering lowImage     
#    :type float 

#    :param highImage: the image to high-pass filter (either greyscale shape=(rows,cols) or colour shape=(rows,cols,channels))     
#    :type numpy.ndarray          
#    :param highSigma: the standard deviation of the Gaussian used for low-pass filtering highImage before subtraction to create the high-pass filtered image     
#    :type float           

#    :returns returns the hybrid image created            
#    by low-pass filtering lowImage with a Gaussian of s.d. lowSigma and combining it with             
#    a high-pass image created by subtracting highImage from highImage convolved with            
#    a Gaussian of s.d. highSigma. The resultant image has the same size as the input images.     
#    :rtype numpy.ndarray    
       
# Your code here.
#   gaussian template creation by calling function makeGaussianKernel
    lowTemplate=makeGaussianKernel(lowSigma)
    highTemplate=makeGaussianKernel(highSigma)
#   smoothning the image
    lowResultantImage=convolve(lowImage,lowTemplate)
#   sharpening the image
    highResultantImage=1-convolve(highImage,highTemplate)
#    superimposing the sharp and smooth image to generate a hybrid
    hybridImage=lowResultantImage+highResultantImage

    return hybridImage
   
def makeGaussianKernel(sigma: float) -> np.ndarray: 
    
#    Use this function to create a 2D gaussian kernel with standard deviation sigma.
#    The kernel values should sum to 1.0, and the size should be floor(8*sigma+1) or      
#    floor(8*sigma+1)+1 (whichever is odd) as per the assignment specification.        
   # Your code here.
#   Determining the value of size by sigma
    size = (int)(8.0*sigma + 1.0)
    pie=math.pi
#   check if size is even 
    if (size%2==0):
        size+=1
#   half value to size
    mid=int(size//2)
#   variable constant stores value 2*sigma^2
    constant=2*(math.pow(sigma,2))
#   define the kernel with dimension size*size
    kernel=np.zeros((size,size))
#   iterating through the dimension, value for each coordinate in the kernel is calculated
    for k in range(0,size):
        for j in range(0,size):
            kernel[k][j]=(math.exp(-((pow(k-mid,2)+pow(j-mid,2))/constant)))/(pie*constant)

    return kernel

                                                              
                                                        
                              
```
